```python
regions=['North America','Europe','Japan','Rest of World']
genres=df['Genre'].unique()
for region in regions:
    print(f"Region: {region}")
    for genre in genres:
        sales=df[df['Genre']==genre][region].sum().round(2)
        print(f"{genre}:{sales}")
    print()
```

```python
#7.Correlatipon through visualization(Using Scatter Plot)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['North America'],df['Europe'])
plt.plot([0,5],[0,5],'k--',label='y=x')
plt.title('Sales VS North America and Europe')
plt.xlabel('North America')
plt.ylabel('Europe')
plt.show()
```

```python
#8.Compare the sales growth trend for "Action" and "Shooter" genres over the years
```

```python
filtered_df = df[df['Genre'].isin(['Action', 'Shooter'])]
sales_by_year_genre = filtered_df.groupby(['Year', 'Genre'])[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
sales_by_year_genre['Global'] = sales_by_year_genre.sum(axis=1)
sales_by_year_genre = sales_by_year_genre.reset_index()
plt.figure(figsize=(12, 6))
for genre in ['Action', 'Shooter']:
    genre_data = sales_by_year_genre[sales_by_year_genre['Genre'] == genre]
    plt.plot(genre_data['Year'], genre_data['Global'], label=genre, marker='o')
plt.title('Sales Growth Trend for "Action" and "Shooter" Genres Over the Years', fontsize=15)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Global Sales', fontsize=10)
plt.show()
```
